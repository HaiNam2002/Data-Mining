```python
# Cài đặt các công cụ để thực hiện thuật toán:
# - pandas: công cụ để làm việc với biểu đồ.
# - seaborn, matplotlib: công cụ để vẽ đồ thị.
# - sklearn: thư viện chứa các thuật toán machine learning.

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# Dùng hàm dưới để đưa dữ liệu vào (có thể đưa csv, excel,...)

data = pd.read_csv(r"D:\Bài Tập Đại Học\KHO\K-Mean data.csv")
data
```

```python
# Gọi và gán biến cho cột dữ liệu mà ta sẽ dùng hoặc muốn xem.
data=data[['precipitation','wind']]
data
```

```python
# Gọi và gán biến x,y vào 2 cột precipitation và wind

x = data['precipitation']
y = data['wind']

# Vẽ đồ thị các điểm trong dữ liệu.
sns.scatterplot(data=data,x='precipitation',y='wind')
plt.title('Seattle Weather')
plt.show()
```

```python
# Zip 2 biến x,y ở trên vào list và gán data_using với list

data_using = list(zip(x,y))
print(data_using)
```

```python
# Thực hiện tính tổng bình phương khoảng cách giữa các điểm, đưa vào list để hỗ trợ cho việc vẽ đồ thị Elbow

Sum_Of_Squared_Distance = []
for k in range(1,11):
    km = KMeans(n_clusters=k,n_init='auto').fit(data_using)
    Sum_Of_Squared_Distance.append(km.inertia_)
    
print(Sum_Of_Squared_Distance)
```

```python
# Vẽ đồ thị Elbow

plt.plot(range(1,11),Sum_Of_Squared_Distance,'bx-')
plt.xlabel('K - Values')
plt.ylabel('Sum Of Squared Distance')
plt.title('Elbow Method')
plt.show()
```

```python
# Chọn k=3 và tiến hành phân cụm bằng thuật toán K-Mean

kmeans = KMeans(n_clusters=3,n_init='auto').fit(data_using)
kmeans.labels_
```

```python
# Hiện điểm trung tâm cuối cùng

centroids=kmeans.cluster_centers_
print(centroids)
```

```python
# Gắn nhãn và thêm cột cluster vào bảng data

data['cluster']=kmeans.labels_
data
```

```python
# Đếm các giá trị trong cluster.

data['cluster'].value_counts()
```

```python
# Vẽ đồ thị phân cụm data bằng phương pháp K-Mean

sns.scatterplot(x='precipitation',y='wind',hue='cluster',data=data)
plt.title('Seattle Weather')
plt.show()
```
